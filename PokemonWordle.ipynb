```python
hiragana = "アイウエオカキクケコサシスセソタチツテトナニヌネノハヒフヘホマミムメモヤユヨラリルレロワヲンガギグゲゴザジズゼゾダヂヅデドバビブベボパピプペポヴァィゥェォッャュョー♀♂2Z・:"
hira = {}
N_ = len(pk)
for i in range(len(hiragana)):
  hira[hiragana[i]] = 0
for i in range(N_):
  temp = []
  for j in range(5):
    temp.append(pk[i][j])
  temp = list(set(temp))
  for j in range(len(temp)):
    hira[temp[j]] += 1
pk_sc = {}
for i in range(N_):
  sum = 0
  temp = []
  for j in range(5):
    temp.append(pk[i][j])
  temp = list(set(temp))
  for j in range(len(temp)):
    sum += hira[temp[j]]
  pk_sc[pk[i]] = sum
# print(sorted(pk_sc.items(), reverse=True, key=lambda x : x[1]))

def search_(N, attempt):
  green = []
  green_ = []
  yellow = [[],[]]
  other = []
  for p in range(len(attempt)):
    length = len(attempt[p])
    letter = 0
    count = 0
    while(letter < length):
      if attempt[p][letter] == "/":
        count += 1
        letter += 1
        green.append([attempt[p][letter], letter - count])
        green_.append(attempt[p][letter])
      elif attempt[p][letter] == "^":
        count += 1
        letter += 1
        yellow[0].append(attempt[p][letter])
        yellow[1].append(letter - count)
      elif not attempt[p][letter] in green_ and not attempt[p][letter] in yellow[0]:
        other.append(attempt[p][letter])
      letter += 1
  # print(green, yellow, other)
  extract = []
  for i in range(N):
    flag = 0
    for j in range(5):
      if pk[i][j] in other:
        flag = 1
        break
    if flag == 0:
      extract.append(pk[i])
  extract1 = []
  if len(green) != 0:
    for i in range(len(extract)):
      flag = 0
      for g in green:
        if g[0] != extract[i][g[1]]:
          flag = 1
          break
      if flag == 0:
        extract1.append(extract[i])
    extract = extract1
  extract1 = []
  if len(yellow[0]) != 0:
    for i in range(len(extract)):
      flag = 0
      for y in range(len(yellow[0])):
        if not yellow[0][y] in extract[i] or yellow[0][y] == extract[i][yellow[1][y]]:
          flag = 1
          break
      if flag == 0:
        extract1.append(extract[i])
    extract = extract1
  result = {}
  for i in range(len(extract)):
    result[extract[i]] = pk_sc[extract[i]]

  return sorted(result.items(), reverse=True, key=lambda x : x[1])
```

```python
import random
def generate_quiz():
  return pk[random.randrange(N_)]
def response(obj, answer):
  temp = []
  s = []
  for j in range(len(answer)):
    s.append(answer[j])
  s = list(set(s))
  count = {}
  for j in range(len(s)):
    count[s[j]] = 0
  for j in range(len(answer)):
    if answer[j] == obj[j]:
      temp.append("/")
    elif obj.count(answer[j]) > count[answer[j]]:
      temp.append("^")
      count[answer[j]] += 1
    else:
      temp.append("*")
  if temp.count("/") == 5:
    return True
  for i in range(5- len(temp)):
    temp.append("*")
  
  # print(temp)
  result = ""
  for j in range(5):
    if temp[j] == "*":
      result = result + answer[j]
    elif temp[j] == "^":
      result = result + "^" + answer[j]
    else:
      result = result + "/" + answer[j]

  return result
```

```python
def solve(n, ini):
  t = []
  for i in range(n):
    step = 1
    obj = generate_quiz()
    answer = ini
    result = []
    while(response(obj, answer) != True):
      result.append(response(obj, answer))
      if len(search_(N_,result)) == 0:
        print(obj, result)
      answer = search_(N_,result)[0][0]
      step += 1
    t.append(step)
    print(result, answer)
  sum = 0
  for i in range(n):
    sum += t[i]
  return sum/n
```
